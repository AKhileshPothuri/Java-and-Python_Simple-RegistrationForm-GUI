```python
from tkinter import*
from tkinter import messagebox 
import xml.etree.cElementTree as ET
import json
root = Tk()
root.geometry('600x600')
root.title("Registration Form")
data={}
def insert(): 
	if str(var.get())=="1":
		gender="Male"
	elif str(var.get())=="2":
		gender="Female"
	elif str(var.get())=="3":
		gender="Other"
	else:
		gender="Not Selected"

	if (entry_1.get() == "" or entry_2.get()=="" or entry_4.get()=="" or gender=="Not Selected" or entry_5.get()=="" or entry_6.get()==""): 

		messagebox.showerror("Error in submission","Please fill all the details")   
	elif (entry_5.get()!=entry_6.get()): 

		messagebox.showerror("Error in submission","Passwords don't match")   

	else:
		jname=entry_1.get()
		jemail=entry_2.get()
		address=entry_4.get()
		password=entry_5.get()
		data[jname] = {'email':jemail, 'gender':gender,'qualification':opt.get(),'password':password,'address':address}
		with open('mydata.json', 'w') as f:
			json.dump(data, f)
		root = ET.Element("root")
		doc = ET.SubElement(root, "doc")

		ET.SubElement(doc, "field1", name="Name").text = jname
		ET.SubElement(doc, "field2", name="Email").text = jemail
		ET.SubElement(doc, "field3", name="Gender").text = gender
		ET.SubElement(doc, "field4", name="Qualification").text =opt.get()
		ET.SubElement(doc, "field5", name="password").text =password
		ET.SubElement(doc, "field6", name="Address").text = address


		tree = ET.ElementTree(root)
		tree.write("mydata.xml")

		messagebox.showinfo("Thankyou for registering!","Details saved successfully")
		entry_1.delete(0,'end')
		entry_2.delete(0,'end')
		entry_4.delete(0,'end')
		entry_5.delete(0,'end')
		entry_6.delete(0,'end')
		var.set(0)
label_0 = Label(root, text="Registration form",width=20,font=("bold", 20))
label_0.place(x=90,y=53)
label_1 = Label(root, text="FullName:",width=20,font=("bold", 10))
label_1.place(x=80,y=130)
entry_1 = Entry(root)
entry_1.place(x=240,y=130)
label_2 = Label(root, text="Email:",width=20,font=("bold", 10))
label_2.place(x=68,y=180)
entry_2 = Entry(root)
entry_2.place(x=240,y=180)

label_3 = Label(root, text="Gender:",width=20,font=("bold", 10))
label_3.place(x=70,y=230)
var = IntVar()
Radiobutton(root, text="Male",padx = 5, variable=var, value=1).place(x=235,y=230)
Radiobutton(root, text="Female",padx = 20, variable=var, value=2).place(x=290,y=230)
Radiobutton(root, text="Other",padx = 25, variable=var, value=3).place(x=370,y=230)
label_7=Label(root, text="Qualification:", width=20, font=("Bold", 10))
label_7.place(x=72, y=280)
opt = StringVar()
choices={'Class X','Class XII','Bachelors Degree','Masters Degree','PhD.'}
opt.set('Class X')
popupmenu = OptionMenu(root,opt,*choices)

popupmenu.place(x=240,y=280)
label_5 = Label(root, text="Enter Password:",width=20,font=("bold", 10))
label_5.place(x=74,y=330)
entry_5 = Entry(root,show="*")
entry_5.place(x=240,y=330)
label_6 = Label(root, text="Confirm Password:",width=20,font=("bold", 10))
label_6.place(x=76,y=380)
entry_6 = Entry(root,show="*")
entry_6.place(x=240,y=380)

label_4 = Label(root, text="Address:",width=20,font=("bold", 10))
label_4.place(x=78,y=430)
entry_4 = Entry(root)
entry_4.place(x=240,y=430)
Button(root, text='Submit',width=20,bg='brown',fg='white',command=insert).place(x=180,y=480)

root.mainloop()



```
